```python
import pandas as pd
```

```python
instancesHeader = ['GRAPH', 'SOURCE', 'TARGET']

instancesDf = pd.read_csv('expInstances.inst', delimiter=' ', header=None, names=instancesHeader)
instancesDf['GRAPH'] = instancesDf['GRAPH'].apply(lambda x : x.split('/')[-1])
instancesDf
```

```python
resultsHeader = ['ALGO', 'HOST', 'GRAPH', 'NODES', 'ARCS', 'SOURCE', 'TARGET', 'ITERATIONS', 'EXTRACTIONS', 'TARGET_SOLUTIONS', 'TIME', 'MEMORY', 'MAX_HEAP_SIZE']

mdaDf = pd.read_csv('mda_exp_3d.csv', delimiter=';', header=None, names=resultsHeader)
martinsDf = pd.read_csv('martins_exp_3d.csv', delimiter=';', header=None, names=resultsHeader)

mergedDf = pd.merge(instancesDf, mdaDf, on=['GRAPH', 'SOURCE', 'TARGET'], how='left')
mergedDf = pd.merge(mergedDf, martinsDf, on=['GRAPH', 'NODES', 'ARCS', 'SOURCE', 'TARGET'], how='left', suffixes=("_mda", "_martins"))
mergedDf
mergedDf.loc[mergedDf['TIME_martins']<0.0001, 'TIME_martins'] = 0.0001
mergedDf.loc[mergedDf['TIME_mda']<0.0001, 'TIME_mda'] = 0.0001
```

```python
mergedDf = mergedDf.loc[mergedDf['TIME_mda'].notna() | mergedDf['TIME_martins'].notna()]
mergedDf.loc[mergedDf['TIME_martins'].isna(), 'TIME_martins'] = 7200
mergedDf.loc[mergedDf['TIME_martins'] > 7200, 'TIME_martins'] = 7200
mergedDf['SOLVED_mda'] = mergedDf['TIME_mda'] < 7200
mergedDf['SOLVED_martins'] = mergedDf['TIME_martins'] < 7200
mergedDf
```

```python
mergedDf['SPEEDUP'] = mergedDf['TIME_martins']/mergedDf['TIME_mda']
#mergedDf['DENSITY'] = mergedDf['DENSITY'].apply(lambda x: round(x))
mergedDf['LABELS_PER_SEC_mda'] = mergedDf['EXTRACTIONS_mda']/mergedDf['TIME_mda']
mergedDf['LABELS_PER_SEC_martins'] = mergedDf['EXTRACTIONS_martins']/mergedDf['TIME_martins']
mergedDf = mergedDf.drop(mergedDf[mergedDf.TARGET ==0].index)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
fig.patch.set_alpha(0.0)
ax1 = fig.add_subplot(111)

c1 = '#FFC857'
c2 = '#255F85'

# print(mergedDf[['GRAPH', 'TARGET', 'ARCS']])
mergedDf['asymptotic_mda'] = mergedDf['EXTRACTIONS_mda'] * np.log(mergedDf['TARGET']) + mergedDf['ARCS']*mergedDf['TARGET_SOLUTIONS_mda']
mergedDf['asymptotic_martins'] = mergedDf['EXTRACTIONS_mda'] * np.log(mergedDf['TARGET']) + mergedDf['ARCS']*mergedDf['TARGET_SOLUTIONS_mda']*mergedDf['TARGET']

martins_df = mergedDf.loc[mergedDf['TIME_martins']<7200]

ax1.scatter(mergedDf['TARGET'], mergedDf['TIME_mda'], s=10, c=c2, marker="s", label='MDA')
ax1.scatter(martins_df['TARGET'], martins_df['TIME_martins'], s=10, c=c1, marker="o", label='Martins')

plt.ylabel('Time [s]')
#plt.xscale('log')
plt.yscale('log')
plt.xlabel('Nodes in graph')
plt.legend()
#ax1.set_xticks(densities)
#ax1.set_yscale('log')
ax1.yaxis.grid(True)
#ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
#ax2.scatter(mergedDf['TARGET'], mergedDf['asymptotic_mda'], s=10, c='red', marker="o", label='Speedup')
#ax2.scatter(mergedDf['TARGET'], mergedDf['asymptotic_martins'], s=10, c='red', marker="o", label='Speedup')
#ax2.legend(loc='lower right')
#ax2.set_ylabel("Speedup")
#ax2.set_yscale('log')

#plt.title(f'EXP instances running times')
plt.savefig(f'./plots/exp_3d_martins_mda_time.svg')
plt.close()
#plt.show()
```

```python
from scipy.stats import gmean
#extractionLimits = [10**i for i in range (9)]
#extractionLimits
print(f"Graph & $N$ & $N_t$ & time mda & time martins & Speedup\\\\")

mergedDf.sort_values(by='GRAPH', inplace=True, ignore_index=True)

for graph, slicedDf in mergedDf.groupby(by='TARGET'):
    #print(len(slicedDf), slicedDf['TIME_martins'])
    gmean_mda = gmean(slicedDf['TIME_mda'])
    gmean_martins = gmean(slicedDf['TIME_martins'])
    #gmeanSpeedup = gmean(slicedDf.loc[slicedDf['TIME_martins']>0.0001]['TIME_martins']/slicedDf.loc[slicedDf['TIME_mda']>0.0001]['TIME_mda'])
    gmeanSpeedup = gmean_martins/gmean_mda
    print(f"EXP{graph+1} ")
    print(f"& {gmean(slicedDf['EXTRACTIONS_mda']):.2f} & {gmean(slicedDf['TARGET_SOLUTIONS_mda']):.2f}")
    print(f"& {gmean_mda:.4f}") 
    print(f"& {gmean_martins:.4f}")
    print(f"& {gmeanSpeedup:.2f}\\\\\n")
```
